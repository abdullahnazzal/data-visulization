```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Hold all Data

```python
df_movies=pd.read_csv("netflix_titles.csv")
df_movies
```

# Filter JUST release_year after then 2000

```python
df_release_year=df_movies["release_year"] >= 2000
df_release_year=df_movies[df_release_year]
df_release_year
```

# Find a mean of math score for ecah type

```python
df_mean=df_movies.groupby("type",as_index=False)["release_year"].mean()
df_mean
```

# As Plot

```python
sns.pointplot(x="type",y="release_year",data=df_mean)
```

# Count number of Movies and TV Show

```python
sns.countplot(x="type",data=df_release_year)
```

# Best Rating in release_year

```python
graph=sns.barplot(x="release_year",y="rating",data=df_movies)
```
